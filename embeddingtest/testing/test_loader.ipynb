```python
import re

def __strip_replace_text(s: str)->str:
        regex = '([^가-힣0-9a-zA-Z])'
        s = re.sub(pattern=regex, repl="", string=s)
        return s


ss = __strip_replace_text(s="가사, 간병 방문 지원사업")

print(ss)
```
